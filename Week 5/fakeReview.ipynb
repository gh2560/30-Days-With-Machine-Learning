```python
import os

data = []

root_dir = r"D:\workspace\dataset\op_spam_v1.4"

for polarity in os.listdir(root_dir):
    polarity_path = os.path.join(root_dir, polarity)
    if not os.path.isdir(polarity_path):
        continue

    for category in os.listdir(polarity_path):
        category_path = os.path.join(polarity_path, category)

        if not os.path.isdir(category_path):
            continue

        label = 1 if "deceptive" in category.lower() else 0

        for file in os.listdir(category_path):
            file_path = os.path.join(category_path, file)

            if not os.path.isfile(file_path):
                continue

            with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                text = f.read()

            data.append({
                "text": text,
                "label": label
            })

print("Total:", len(data))
```
