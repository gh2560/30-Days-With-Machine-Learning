```python
import kagglehub
import numpy as np
import seaborn as sns
import pandas as pd
import os
path = kagglehub.dataset_download("uciml/pima-indians-diabetes-database")
print(path)

print(os.listdir(path))
file_path = os.path.join(path, "diabetes.csv")
print(file_path)
df=pd.read_csv(file_path)
```

```python
df.head()
```

```python
#data pre-process
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
print(df.isna().sum())
#split prediction and information
y=df["Outcome"]
x=df.drop(columns="Outcome")

cat_cols=df.select_dtypes(include='object').columns.tolist()
num_cols=df.select_dtypes(include='number').columns.tolist()
print(cat_cols,num_cols)
#all data is labeled and only contain number data
#don't need to impute/encode
x_train,x_val,y_train,y_val=train_test_split(x,y,train_size=0.8,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_absolute_error,precision_score
model=LogisticRegression(random_state=0)
model.fit(x_train,y_train)
predictions=model.predict(x_val)
mae=mean_absolute_error(y_val,predictions)
score=precision_score(y_val,predictions)
print(mae)
print(score)
```
