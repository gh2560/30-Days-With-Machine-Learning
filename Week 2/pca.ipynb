```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Tạo một DataFrame giả định
data = {
    'Feature1': [2.5, 0.5, 2.2, 1.9, 3.1],
    'Feature2': [2.4, 0.7, 2.9, 2.2, 3.0],
    'Feature3': [1.2, 0.4, 1.8, 1.7, 2.2],
    'Feature4': [3.5, 1.2, 2.8, 2.6, 3.7]
}

df = pd.DataFrame(data)

# Chuẩn hóa dữ liệu trước khi áp dụng PCA
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Áp dụng PCA và giảm chiều dữ liệu xuống 2
pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_scaled)

# Chuyển đổi kết quả thành DataFrame
df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2'])

# In kết quả
print("Dữ liệu sau PCA:")
print(df_pca)

# In các Eigenvalues (biến thiên giải thích)
print("\nEigenvalues (Giải thích biến thiên):")
print(pca.explained_variance_)

# In các Eigenvectors (Hệ số của PCs)
print("\nEigenvectors (Hệ số của PCs):")
print(pca.components_)

# Tính tỷ lệ giải thích của từng thành phần chính (PC)
explained_variance_ratio = pca.explained_variance_ratio_
print("\nTỷ lệ giải thích biến thiên:")
print(explained_variance_ratio)
```
